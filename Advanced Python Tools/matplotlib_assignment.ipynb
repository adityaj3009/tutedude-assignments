```python
import numpy as np 
import matplotlib.pyplot as plt
# 1. Create a simple line plot for the following data:
# x = [1, 2, 3, 4, 5]
# y = [10, 15, 25, 30, 50]
# a. Plot the data.  
# b. Customize the plot by adding a title, axis labels, and a grid. 

x = np.array([1,2,3,4,5])
y = np.array([10,15,25,30,50])

plt.title("Line plot")
plt.grid(True) # this adds grid 
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x,y)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

# 2. Create a bar graph to represent the marks scored by students in a subject:  
#students = ['John', 'Jane', 'Alice', 'Bob']
#marks = [75, 85, 60, 90]
#a. Plot the data as a bar graph.  
#b. Customize the colors and add a title.  

students = np.array(['John', 'Jane', 'Alice', 'Bob'])
marks = np.array([75, 85, 60, 90])
plt.title("Marks scored by students", fontdict={'fontsize': 20, 'fontweight':5, 'color': 'black'})
plt.bar(students,marks, width=0.5, align='center', color = "red")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 3. Create a pie chart to represent the percentage distribution of a companys revenue from different regions:  
# regions = ['North America', 'Europe', 'Asia', 'Others']
# revenue = [45, 25, 20, 10]
# a. Create a pie chart with the region names as labels.  
# b. Highlight the region with the highest revenue.  

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# 0.1 - first slice, 0 - other remaining slice
explode = [0.1, 0, 0, 0]  # Highlight North America, shows much much slice should pop outs

plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%') # shows percentages on each slice
plt.title("Revenue Distribution by Region")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 4. Generate a histogram to show the frequency distribution of 
# randomly generated integers between 1 and 100 (sample size = 1000).  

data = np.random.randint(1, 101, 1000)

# bins=10 -  range of value divided in 10 parts
plt.hist(data, bins=10, color='blue', edgecolor='black')

plt.title("Histogram of Random Integers (1 to 100)")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

plt.show()
```
