```python
# librabries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
dataset.dtypes
```

```python
dataset.shape
```

```python
# delete useless column
df = dataset.drop(["Lat","Long", "Province/State"],axis=1, inplace=True)

df
```

```python
dataset.head()
```

```python
# aggregate rows by country
corona_dataset_aggregated= dataset.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# visualize this data related to country
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
corona_dataset_aggregated.loc["Qatar"].plot()
corona_dataset_aggregated.loc["China"].plot()

plt.legend()

```

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
# calculate first derivative of curve
corona_dataset_aggregated.loc["India"].diff().plot()
```

```python
# max infection rate
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
# Calculate max infection rates
max_infection_rates = []
for country in corona_dataset_aggregated.index:
    daily_increases = corona_dataset_aggregated.loc[country].diff().max()
    max_infection_rates.append(daily_increases)

corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```

```python
# importing datset 2
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
print(happiness_report)
```

```python
# drop useless column
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report.drop(useless_cols, axis =1, inplace=True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region", inplace=True)
happiness_report.head()
```
